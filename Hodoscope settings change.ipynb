```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline

from HodoUtils import HodoSummary
```

```python
s1 = HodoSummary('./data/3712/ftHodo_OverviewFile_-1073741568_03-28-2018_11.41.35_AM.txt')
s2 = HodoSummary('./data/3712/nsb02nsa12/ftHodo_OverviewFile_-1073741568_03-28-2018_12.36.33_PM.txt')
```

```python
x1, y1, yerr1 = s1.get_sector_data(1, 1, "m_tiles_charge")
x2, y2, yerr2 = s2.get_sector_data(1, 1, "m_tiles_charge")

fig, ax = plt.subplots(1, 1)
ax.errorbar(x1, y1, marker='o', linestyle='None', yerr=yerr1)
ax.errorbar(x2, y2, marker='o', linestyle='None', yerr=yerr2)
ax.grid()
ax.set_title('Plot of the same components (sector 1, layer 1), analysed '
             '\nwith different settings, using the same data-set')
ax.set_ylabel('Max Charge')
ax.set_xlabel('Component')
plt.show()
```

```python
# A plot of the difference between the charge value for each component between the 2 summary files

y1 = s1.get_data_flat("m_tiles_charge")
y2 = s2.get_data_flat("m_tiles_charge")
dy = y1 - y2

fig, ax = plt.subplots(1, 1)
ax.plot(dy, marker='.', linestyle='None')
ax.grid()
ax.set_ylabel('delta charge')
ax.set_xlabel('Components (all)')
ax.set_yscale('log')
plt.show()
```
